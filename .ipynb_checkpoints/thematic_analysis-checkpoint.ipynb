```python
# read the csv data
df_sentiment = pd.read_csv('./data/reviews_with_sentiment.csv')
df_sentiment.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
```

```python
# extract the key words from the sentiment analysis
keywords = extract_keywords(df_sentiment)
print("Extracted keywords:", keywords)
```

```python
# manual theme mapping based on the dataset
theme_mapping = {
    'Performance Issues': [
        'slow', 'crash', 'crashes', 'loading', 'lag', 'freeze', 'frozen',
        'hang', 'delay', 'unresponsive'
    ],
    'Account Access Issues': [
        'login', 'log in', 'password', 'authentication', 'fingerprint',
        'signin', 'sign in', 'otp', 'verification', 'access', 'logout'
    ],
    'User Interface & Experience': [
        'ui', 'interface', 'user interface', 'ux', 'navigation',
        'design', 'layout', 'menu', 'user friendly', 'friendly', 'intuitive'
    ],
    'Customer Support': [
        'support', 'customer support', 'help', 'customer service', 'call center',
        'agent', 'response', 'respond', 'staff', 'contact'
    ],
    'Feature Requests & Suggestions': [
        'feature', 'features', 'update', 'improve', 'improvement', 'upgrade',
        'suggestion', 'add', 'added', 'new option', 'option', 'enhancement'
    ],
    'Connectivity Issues': [
        'network', 'internet', 'connection', 'offline', 'data', 'connectivity'
    ],
    'Financial Transactions': [
        'transaction', 'transfer', 'money', 'payment', 'send', 'receive',
        'balance', 'statement', 'limit', 'charge', 'fee'
    ],
    'Security & Privacy Concerns': [
        'security', 'secure', 'privacy', 'safe', 'protection', 'hack', 'fraud', 'scam'
    ]
}
```

```python
# based on the above theme we map manually
df_themed = assign_themes(df_sentiment, theme_mapping)
```
